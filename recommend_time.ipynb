```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import logging

# 경고 메시지를 출력하지 않도록 설정
logging.getLogger("transformers").setLevel(logging.ERROR)

class MealTimePredictor:
    def __init__(self):
        self.model_name = "gpt2"  # GPT-2 모델 사용
        self.tokenizer = GPT2Tokenizer.from_pretrained(self.model_name)
        self.model = GPT2LMHeadModel.from_pretrained(self.model_name)

        # 아침, 점심, 저녁에 관련된 단어들
        self.morning_words = morning_data
        self.lunch_words = lunch_data
        self.dinner_words = dinner_data
        
    def predict_meal_time(self, input_text):
        # 입력 텍스트에 대한 토큰화
        input_ids = self.tokenizer.encode(input_text, return_tensors="pt")

        # GPT-2 모델로 예측
        with torch.no_grad():
            output = self.model.generate(input_ids, max_length=50)

        # 예측 결과를 텍스트로 디코딩
        predicted_text = self.tokenizer.decode(output[0], skip_special_tokens=True)

        # 아침, 점심, 저녁 중 어떤 시간대에 관련이 있는지 확인
        if any(word in predicted_text for word in self.morning_words):
            return "아침"
        elif any(word in predicted_text for word in self.lunch_words):
            return "점심"
        elif any(word in predicted_text for word in self.dinner_words):
            return "저녁"
        else:
            return "시간대를 알 수 없음"

# 모델 인스턴스 생성
meal_time_predictor = MealTimePredictor()
```

```python
# 텍스트 입력
user_input = input("오늘의 활동이나 느낌을 입력해주세요: ")

# 모델을 사용하여 예측
predicted_meal_time = meal_time_predictor.predict_meal_time(user_input)

# 결과 출력
print("입력 텍스트:", user_input)
print("예측된 시간대:", predicted_meal_time)
```
